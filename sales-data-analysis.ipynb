```python
ax = sns.countplot(data=data,x='Product')
plt.xticks(rotation=90)
for bars in ax.containers:
    ax.bar_label(bars)
```

# **Most Sold Product**

```python
plt.figure(figsize=(20,5))
sales = data.groupby(['Product'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
sns.barplot(data=sales,x='Product',y='Sales')
plt.xticks(rotation=70)
```

***Macbook*** pro laptop is the most sold product followed by ***iphone*** whereas ***AAA Batteries(4-pack)*** is the least sold product by comparison.

# **Sales Per City**

```python
plt.figure(figsize=(20,5))
sales = data.groupby(['City'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
sns.barplot(data=sales,x='City',y='Sales')
plt.xticks(rotation=70)
```

**San Francisco** tops the chart with over 8M sales whereas **Austin** is the least city in sales.

```python
data['QuantitySold'] = data.groupby('Quantity Ordered')['Quantity Ordered'].sum()
print(f"Quantity Sold: {data['Quantity Ordered'].sum()}")
```
